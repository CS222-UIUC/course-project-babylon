```python
%matplotlib inline
import pandas as pd
import mplfinance as mpf
from datetime import datetime
import pandas_datareader as web
import plotly.graph_objs as go
import plotly
```

```python
#https://www.youtube.com/watch?v=GcXgxfbcFrQ
file = "AMZN.csv"
data = pd.read_csv(file)
data.columns
data.Date = pd.to_datetime(data.Date)
#data.info()
data = data.set_index('Date')
#mpf.plot(data)
mpf.plot(data['2022-03':'2022-06'],figratio =(20,12), type = 'candle',
         title = 'Amazon Price 2022',mav = (20),volume=True,tight_layout = True,style = 'yahoo')
```

```python
#https://www.youtube.com/watch?v=4fhBXFSS1lc
df= pd.read_csv( "AMZN.csv")
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
figure = go.Figure(data = [
    go.Candlestick(x=df.index,low = df['Low'],high = df['High'],open=df['Open'],close = df ['Close'])
])
# df = df.iloc[::1]
# # df['date'] = pd.to_datetime(df['date'])
# #
# df.Date = pd.to_datetime(df.Date)
# data = df.set_index('Date')
# #
figure.update_layout(
    title = 'Amazon Price',
    yaxis_title = 'Amazon Stock Price USD ($)',
    xaxis_title = 'Date'
)
figure.show()
```
